```python
import socket

target = input("Enter IP Address: ")

print("Scanning...")

for port in range(20, 31):
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.settimeout(1)

    result = s.connect_ex((target, port))

    if result == 0:
        print("Port", port, "is OPEN")

    s.close()
```
