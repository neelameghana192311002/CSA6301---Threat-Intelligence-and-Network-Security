```python
from collections import Counter

# Create a sample log file
with open("log.txt", "w") as file:
    file.write("""192.168.1.10
192.168.1.10
192.168.1.10
192.168.1.20
192.168.1.10
192.168.1.30
""")

# Read the log file
with open("log.txt", "r") as file:
    ips = file.read().splitlines()

# Count IP occurrences
count = Counter(ips)

# Detect possible DoS attack
for ip, visits in count.items():
    if visits > 3:
        print(ip, "- Possible DoS Attack")
    else:
        print(ip, "- Normal Traffic")
```
