```python
import hashlib

# Function to calculate SHA-256 hash of a file
def hash_file(path):
    with open(path, "rb") as f:
        return hashlib.sha256(f.read()).hexdigest()


# Create sample files
with open("original.txt", "w") as f:
    f.write("System32 backup data")

with open("copy.txt", "w") as f:
    f.write("System32 backup data")

with open("tampered.txt", "w") as f:
    f.write("System32 backup data - modified")


# List of files
files = ["original.txt", "copy.txt", "tampered.txt"]

# Calculate hashes
hashes = {file: hash_file(file) for file in files}

# Original file hash
baseline = hashes["original.txt"]

print("Baseline (original.txt):", baseline)

# Compare files with the original
for file in files[1:]:
    if hashes[file] == baseline:
        status = "IDENTICAL"
    else:
        status = "MODIFIED/TAMPERED"

    print(f"{file}: {status}")
```
